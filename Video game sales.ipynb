```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("video_games_sales.csv")
```

```python
#We see that "Publisher" and "Developer" features have same values so dropping the "Developer" feature:
df.drop(["Developer"],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Making another dataframe with null values being dropped
df_mod=df.dropna()
```

```python
df_mod.isnull().sum()
```

```python
df_mod.info()
```

```python
#Change Year of Release feature's dtype to int
df_mod["Year_of_Release"]=df_mod["Year_of_Release"].astype("int")
```

```python
df_mod.info()
```

```python
#Barplot on the basis of "Year of Release"
plt.figure(figsize=(12,6))
grp_year=df_mod.groupby(['Year_of_Release']).sum()
sns.barplot(x=grp_year.index,y=grp_year["Global_Sales"],color="cyan")
plt.title('Global Games sales per Year in Million Units')
plt.ylabel('Sales in Millions')
plt.show()

#we see that games sales peeked in 2008 and since been on a decline
```

```python
plt.figure(figsize=(12,6))
grp_platform=df_mod.groupby(['Platform']).sum()
grp_platform=grp_platform.sort_values("Global_Sales")
sns.barplot(x=grp_platform.index,y=grp_platform["Global_Sales"],color="cyan")
plt.title('Global Sales By Platform in Millions')
plt.show()
```

```python
#Using subplots to check different region sales
JPN=grp_year['JP_Sales']
NA=grp_year['NA_Sales']
EU=grp_year['EU_Sales']
Other=grp_year['Other_Sales']
grp_year_features=[JPN,NA,EU,Other]

plt.figure(figsize=(15,8))
plt.style.use('fivethirtyeight')
sns.set_context(font_scale=1,rc={"font.size":12,"axes.labelsize":15})

for i in enumerate(grp_year_features):
    plt.subplot(2,2,i[0]+1)
    plt.subplots_adjust(hspace=0.5)
    sns.barplot(x=grp_year.index,y=i[1],color="red")    
    plt.xticks(rotation=90)
 
```

# Lets look at Genre popularity by Region:

```python
grp_genre_region=df_mod.groupby(['Genre']).sum().drop(['Critic_Score','Critic_Count','User_Count','Year_of_Release','Global_Sales'],axis=1)
grp_genre_region
```

```python
grp_genre_region.plot.bar(stacked=True,figsize=(15,8))
plt.title('Game Sales by Region and Genre')
plt.show()
```

```python
##All we can see here is that NA and EU region sales are higher than the rest. Action, RPG, Shooter, Sports genres are the most popular in all regions.
```

# Lets calculate percentage for a more practical view of Region and Genre sales:

```python
np.array(grp_genre_region.sum())
```

```python
#adding "Total" row
grp_genre_region.loc['Total'] = [2692.35, 1611.31,  437.88,  564.27]
grp_genre_region
```

```python
#Calculating the percentages for each column
grp_genre_region_pct=pd.DataFrame([])
for col in grp_genre_region.columns:
    grp_genre_region_pct[col]=grp_genre_region[col].apply(lambda x:(x/grp_genre_region[col]['Total'])*100)
```

```python
grp_genre_region_pct
```

```python
grp_genre_region_pct.stack().unstack(level=0).drop(['Total'],axis=1).plot.bar(stacked=True,figsize=(15,8),cmap='Paired')
plt.legend(frameon=False,bbox_to_anchor=[0,0,1.1,0.95])
plt.title('Sales by Region and Genre ( in %)')
plt.show()
```

```python
##We can see that NA and EU have pretty similar preferences whereas JP has different. JP prefers Role-Playing genre the most.
```
